```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lab_data = pd.read_csv('Loc1_session1_group1_2024-04-12_16-14-40.csv')
lab_data['Time'] = lab_data['Time']-lab_data['Time'].loc[0]
data_part1 = lab_data.loc[lab_data['Time']<315] #Near Jacobs
data_part2 = lab_data.loc[lab_data['Time']>=315] #Near Soda 
```

```python
#3a: Radiation Data 
counts_data_part1 = data_part1['counts']

plt.plot(data_part1['Time'], counts_data_part1)
plt.xlabel('time')
plt.ylabel('counts')
plt.title('Counts/Time Near Jacobs')
plt.show()
counts_data_part2 = data_part2['counts']

plt.plot(data_part2['Time'], counts_data_part2)
plt.xlabel('time')
plt.ylabel('counts')
plt.title('Counts/Time Near Soda')
plt.show()
```

```python
#3a: Temperature Data 
Temps_data_part1 = data_part1['Temperature']

plt.plot(data_part1['Time'], Temps_data_part1)
plt.xlabel('Time')
plt.ylabel('Temp')
plt.title('Temp/Time Near Jacobs')
plt.show()

Temps_data_part2 = data_part2['Temperature']
plt.plot(data_part2['Time'], Temps_data_part2)
plt.xlabel('Time')
plt.ylabel('Temp')
plt.title('Temp/Time Near Soda')
plt.show()
```

```python
#3a: PM2.5 Data 
PM25_data_part1 = data_part1['PM25']

plt.plot(data_part1['Time'], PM25_data_part1)
plt.xlabel('Time')
plt.ylabel('PM 2.5')
plt.title('PM 2.5/Time Near Jacobs')
plt.show()

PM25_data_part2 = data_part2['PM25']
plt.plot(data_part2['Time'], PM25_data_part2)
plt.xlabel('Time')
plt.ylabel('PM 2.5')
plt.title('PM 2.5/Time Near Soda')
plt.show()
```

```python
#3b: Radiation Data
plt.hist(data_part1['counts'])
plt.title("Histogram of Counts near Jacobs")
plt.xlabel("Counts")
plt.ylabel("Count")
plt.show()
counts_mean_part1 = np.mean(counts_data_part1)
counts_std_part1 = np.std(counts_data_part1)
print("The mean of the counts near Jacobs is", counts_mean_part1)
print("The standard deviation of the counts near Jacobs is", counts_std_part1)
plt.hist(data_part2['counts'])
plt.title("Histogram of Counts near Soda")
plt.xlabel("Counts")
plt.ylabel("Count")
plt.show()
counts_mean_part2 = np.mean(counts_data_part2)
counts_std_part2 = np.std(counts_data_part2)
print("The mean of the counts near Soda is", counts_mean_part2)
print("The standard deviation of the counts near Soda is", counts_std_part2)
```

```python
#3b: Temperature Data
plt.hist(data_part1['Temperature'])
plt.title("Histogram of Temperature near Jacobs")
plt.xlabel("Temperature")
plt.ylabel("Counts")
plt.show()
Temp_mean_part1 = np.mean(Temps_data_part1)
Temp_std_part1 = np.std(Temps_data_part1)
print("The mean of the temperatures near Jacobs is", Temp_mean_part1)
print("The standard deviation of the temperatures near Jacobs is", Temp_std_part1)
plt.hist(data_part2['Temperature'])
plt.title("Histogram of Temperature near Soda")
plt.xlabel("Temperature")
plt.ylabel("Counts")
plt.show()
Temp_mean_part2 = np.mean(Temps_data_part2)
Temp_std_part2 = np.std(Temps_data_part2)
print("The mean of the Temperatures near Soda is", Temp_mean_part2)
print("The standard deviation of the temperatures near Soda is", Temp_std_part2)
```

```python
#3b: PM2.5 Data
plt.hist(data_part1['PM25'])
plt.title("Histogram of PM 2.5 near Jacobs")
plt.xlabel("PM 2.5")
plt.ylabel("Counts")
plt.show()
PM25_mean_part1 = np.mean(PM25_data_part1)
PM25_std_part1 = np.std(PM25_data_part1)
print("The mean of PM2.5 near Jacobs is", PM25_mean_part1)
print("The standard deviation of PM 2.5 near Jacobs is", PM25_std_part1)
plt.hist(data_part2['PM25'])
plt.title("Histogram of PM 2.5 near Soda")
plt.xlabel("PM 2.5")
plt.ylabel("Counts")
plt.show()
PM25_mean_part2 = np.mean(PM25_data_part2)
PM25_std_part2 = np.std(PM25_data_part2)
print("The mean of PM2.5 near Jacobs is", PM25_mean_part2)
print("The standard deviation of PM 2.5 near Jacobs is", PM25_std_part2)
```
