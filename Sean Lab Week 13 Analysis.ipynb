```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = pd.read_csv('Loc1_session1_group1_2024-04-12_16-14-40.csv')
```

```python
#Temperature:
df_Temperature = df['Temperature']

#b.
plt.hist(df_Temperature)
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.title('Occurances in Temperature throughout experiment')

mean_Temperature = np.mean(df_Temperature)
std_Temperature = np.std(df_Temperature)

print(f'Temperature Mean = {mean_Temperature}')
print(f'Temperature std = {std_Temperature}')

#c. 
sum_xi_x_2_Temperature = []
for i in range(len(df_Temperature)):
    sum_xi_x_2_Temperature_i = (df_Temperature[i] - mean_Temperature)**2
    sum_xi_x_2_Temperature.append(sum_xi_x_2_Temperature_i)
sum_xi_x_2_Temperature = np.sum(sum_xi_x_2_Temperature)

variance_Temperature = np.sqrt(sum_xi_x_2_Temperature/(len(df_Temperature)-1))
print(f'Temperature variance = {variance_Temperature}')

#d.
N = np.sum(df_Temperature)
N_samples = len(df_Temperature)
uncertainty = np.sqrt(N)/np.sqrt(2*np.sqrt(N_samples))
print(f'Uncertainty = {uncertainty}')
```

```python
#a. 
plt.plot(df_Temperature)
plt.xlabel('Time (5 seconds)')
plt.ylabel('Temperature (C)')
plt.title('Temperature vs. Time')
```
